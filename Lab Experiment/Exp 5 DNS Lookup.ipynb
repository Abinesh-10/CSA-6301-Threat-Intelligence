```python
# Experiment 5: DNS Lookup (OSINT Basics)

import socket

domains = [
    "www.google.com",
    "www.python.org",
    "notarealdomain12345.com"
]

for domain in domains:
    try:
        ip = socket.gethostbyname(domain)
        print(f"{domain:30s} -> {ip}")
    except socket.gaierror:
        print(f"{domain:30s} -> Could not resolve (invalid/unreachable)")
```
