```python
# Experiment 4: Phishing URL Detector

import re


def check_url(url):
    reasons = []

    # Raw IP address
    if re.match(r"https?://\d{1,3}(\.\d{1,3}){3}", url):
        reasons.append("Uses raw IP address instead of a domain name")

    # @ symbol
    if "@" in url:
        reasons.append("Contains '@' symbol (can hide real destination)")

    # Too many hyphens
    if url.count("-") > 3:
        reasons.append("Too many hyphens in domain")

    # Sensitive keywords without HTTPS
    if any(word in url.lower() for word in ["login", "verify", "update", "secure"]) and not url.startswith("https://"):
        reasons.append("Suspicious keyword without HTTPS")

    return reasons


urls = [
    "https://www.google.com",
    "http://192.168.10.5/login",
    "http://paypal-verify-account.com@evil.com"
]


for url in urls:
    issues = check_url(url)

    verdict = "SUSPICIOUS" if issues else "Looks OK"

    print(f"\nURL: {url}")
    print("Verdict:", verdict)

    for issue in issues:
        print("-", issue)
```
